1. WAP using pandas to remove missing values using data.csv where
    a. purch_amt using median
    b. sale_amt using mean
    c. ord_date using forward Fill
    d. salesman_id using either forward fill and backward fill
    e. ord_no using backward fill

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("new.csv")
df.head()
```

## a)

```python
df['purch_amt'] = df['purch_amt'].fillna(df['purch_amt'].median())
df
```

## b)

```python
df['sale_amt'] = df['sale_amt'].fillna(df['sale_amt'].mean())
df
```

## c)

```python
df['ord_date'] = df['ord_date'].fillna(method='ffill')
df
```

## d)

```python
df['salesman_id'] = df['salesman_id'].fillna(method='ffill')
df
```

## e)

```python
df['ord_no'] = df['ord_no'].fillna(method='bfill')
df
```
